```python
d = {'coll':[1,2,4,3,2,1,6,7,3,9],'col2':[5,8,9,10,12,8,6,2,8,7]}
df2=pd.DataFrame(d)
df2
```

```python
fig = plt.figure()
ax=fig.add_subplot(1,1,1)

ax.plot(df2['coll'],'k.',label='coll')
ax.plot(df2['col2'],'k--',label='col2')
ax.legend(loc='best')
```

```python
#Ticks,labels and legends
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(randn(1000).cumsum())
```

```python
ticks = ax.set_xticks([0,250,500,750,1000])

labels=ax.set_xticklabels(['one','two','three','four','five'],rotation=30,fontsize='small')
ax.set_title('My First Matplotlip plot with Ticks')
ax.set_xlabel('stages')
ax.set_ylabel('Values')
fig
```

```python
props = {'title':'My First matplotlip Plot with Ticks',
        'xlabel':'stages',
        'ylabel':'Values'}

ax.set(**props)
fig
```

```python
fig = plt.figure()
ax=fig.add_subplot(1,1,1)

ax.plot(randn(1000).cumsum(),'k',label='one')
ax.plot(randn(1000),'k--',label='two')
ax.plot(5 * randn(1000),'k.',label='three')

ax.legend(loc='best')
```
